## Ejemplo if else

```python
# 2. Ejemplo de if-else con operadores lógicos y relacionales
# Determina si una persona tiene acceso a un descuento especial basándose en dos criterios combinados

edad = 25
tiene_membresia = True

# La condición requiere que ambas partes sean verdaderas
if edad >= 18 and tiene_membresia:
    print("Acceso al descuento de socio concedido.")
else:
    print("No cumple con los requisitos para el descuento.")
```

## Ejemplo de if else

```python
# 3. Ejemplo de if-elif-else con lógica compleja (Año Bisiesto)
# Este ejemplo es ideal para observar cómo se mezclan todos los tipos de operadores en una estructura múltiple

año=int(input("Ingrese el año: "))
# Un año es bisiesto si es divisible por 4 y no por 100, o si es divisible por 400
if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
    print(f"{año} es un año bisiesto.")
elif año < 0:
    print("Error: El año no puede ser negativo.")
else:
    print(f"{año} no es un año bisiesto.")
```

## Ejemplo de if elif else

```python
# 4. Ejemplo de if-elif-else para cálculo de rangos (Descuentos)
# Este modelo evalúa en qué escala de cantidad se encuentra una compra para asignar un descuento proporcional

cantidad=int(input("Ingrese la cantidad de la compra: "))
if cantidad >= 20:
    descuento = 0.3  # 30% de descuento
elif cantidad >= 10:
    descuento = 0.15 # 15% de descuento
else:
    descuento = 0    # Sin descuento

print(f"Tu descuento aplicado es del {descuento * 100}%.")
```
